```r
players_data_tidy_split<-initial_split(players_data_tidy, prop=0.75, strata=played_hours)

players_data_tidy_train<-training(players_data_tidy_split)

players_data_tidy_test<-testing(players_data_tidy_split)
```

```r
players_data_tidy_recipe<-recipe(played_hours~age, data=players_data_tidy_train)|>
                          step_scale(all_predictors())|>
                          step_center(all_predictors())
players_data_tidy_recipe

players_data_tidy_spec<-nearest_neighbor(weight_func="rectangular", neighbors=tune())|>
                        set_engine("kknn")|>
                        set_mode("regression")
players_data_tidy_spec
```

```r
players_data_tidy_vfold<-vfold_cv(players_data_tidy_train, v=5, strata=played_hours)

players_data_tidy_workflow<-workflow()|>
                            add_recipe(players_data_tidy_recipe)|>
                            add_model(players_data_tidy_spec)
```

```r
grid_vals<-tibble(neighbors=seq(from=1,to=100,by=10))

players_data_tidy_results<-players_data_tidy_workflow|>
                           tune_grid(resamples=players_data_tidy_vfold, grid=grid_vals)|>
                           collect_metrics()|>
                           filter(.metric=="rmse")
players_data_tidy_results
```

```r
players_data_tidy_min<-players_data_tidy_results|>
                       filter(mean==min(mean))
players_data_tidy_min
```

```r
kmin<-players_data_tidy_min|>
      pull(neighbors)

players_data_tidy_new_spec<-nearest_neighbor(weight_func="rectangular", neighbors=kmin)|>
                           set_engine("kknn")|>
                           set_mode("regression")

players_data_tidy_fit<-workflow()|>
                       add_recipe(players_data_tidy_recipe)|>
                       add_model(players_data_tidy_new_spec)|>
                       fit(data=players_data_tidy_train)

players_data_tidy_summary<-players_data_tidy_fit|>
                           predict(players_data_tidy_test)|>
                           bind_cols(players_data_tidy_test)|>
                           metrics(truth=played_hours, estimate=.pred)|>
                           filter(.metric=="rmse")
players_data_tidy_summary
```
